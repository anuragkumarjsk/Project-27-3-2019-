```python
import cv2
import matplotlib.pyplot as plt
#%matplotlib.inline
img1=cv2.imread('/root/Desktop/python/parrot.jpeg')
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2=cv2.imread('/root/Desktop/python/parrothue.jpeg')
merged=cv2.addWeighted(src1=img1,alpha=0.3,src2=img2,beta=0.7,gamma=0)
plt.imshow(merged)
```
