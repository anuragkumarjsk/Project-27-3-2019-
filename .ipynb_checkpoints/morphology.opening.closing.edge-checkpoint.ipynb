```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
%matplotlib.inline
```

```python
def load_img():
    blank_img=np.zeros((600,600))
    font=cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img,text='abcde',org=(50,300),fontFace=font,fontScale=6,color=(255,255,255),thickness=10)
    return blank_img
```

```python
def img_show(img):
    fig=plt.figure(figsize=(15,15))
    aa=fig.add_subplot(111)
    aa.imshow(img,cmap='gray')
```

```python
image=load_img()
img_show(image)
```
