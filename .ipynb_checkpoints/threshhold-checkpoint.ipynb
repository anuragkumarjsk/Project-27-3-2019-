```python
import matplotlib.pyplot as plt
import cv2
%matplotlib.inline
```

```python
img=cv2.imread('/root/Desktop/python/condor.jpeg',0)
plt.imshow(img,cmap='gray')
```

```python
ret,thresh1=cv2.threshold(img,107,255,cv2.THRESH_BINARY_INV)#or use without INV
```

```python
ret
```

```python
plt.imshow(thresh1,cmap='gray')
```

```python
ret,thresh2=cv2.threshold(img,107,255,cv2.THRESH_TRUNC)
plt.imshow(thresh1,cmap='gray')
```

```python
ret,thresh2=cv2.threshold(img,107,255,cv2.THRESH_TOZERO)
plt.imshow(thresh1,cmap='gray')
```

```python
ret,thresh2=cv2.threshold(img,107,255,cv2.THRESH_TOZERO_INV)
plt.imshow(thresh1,cmap='gray')
```

```python
def img_show(img):
    fig=plt.figure(figsize=(15,15))
    aa=fig.add_subplot(111)
    aa.imshow(img,cmap='gray')
```

```python
img_show(img)
```
