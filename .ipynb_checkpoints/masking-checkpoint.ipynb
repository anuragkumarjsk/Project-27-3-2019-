```python
import cv2
import matplotlib.pyplot as plt

img1=cv2.imread('/root/Desktop/python/condor.jpeg')
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2=cv2.imread('/root/Desktop/python/raven.jpeg')
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
img2=cv2.resize(img2,(122,125))
large_image=img1
small_image=img2
```

```python
plt.imshow(large_image)
```

```python
plt.imshow(small_image)
```

```python
xoffset=large_image.shape[1]-small_image.shape[1]
yoffset=large_image.shape[0]-small_image.shape[0]
rows,columns,channels=small_image.shape
```

```python
roi=large_image[yoffset:151,xoffset:333]
plt.imshow(roi)
```

```python
grayimg2=cv2.cvtColor(small_image,cv2.COLOR_RGB2GRAY)
plt.imshow(grayimg2,cmap='gray')
```

```python
mask_inv=cv2.bitwise_not(grayimg2)
```

```python
plt.imshow(mask_inv ,cmap='gray')
```

```python
mask_inv.shape
```

```python
import numpy as np
white_background = np.full(small_image.shape,255,dtype=np.uint8)
bk=cv2.bitwise_or(white_background,white_background,mask=mask_inv)
plt.imshow(bk)
```

```python
fg=cv2.bitwise_or(small_image,small_image,mask=mask_inv)
plt.imshow(fg)
```

```python
final_roi=cv2.bitwise_or(roi,fg)
plt.imshow(final_roi)
```

```python
large_image[yoffset:151,xoffset:333]=final_roi
plt.imshow(large_image)
```
