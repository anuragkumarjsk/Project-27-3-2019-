```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def disp(img,cmap='gray'):
    fig=plt.figure(figsize=(12,10))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
#median blur
#greyscale
#binarythreshhold
#contours

brdgrp=cv2.imread('/root/Desktop/python/manyparrots.jpeg')
disp(brdgrp)
```

```python
sep_brd=cv2.medianBlur(brdgrp,5)
disp(sep_brd)
```

```python
grybrdgrp=cv2.cvtColor(sep_brd,cv2.COLOR_BGR2GRAY)
disp(grybrdgrp)
```

```python
ret,sep_thresh=cv2.threshold(grybrdgrp,40,255,cv2.THRESH_BINARY_INV)
disp(sep_thresh)
```

```python
contours,hierarchy=cv2.findContours(sep_thresh.copy(),cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
```

```python
contours
```

```python
for i in range(len(contours)):
    if hierarchy[0][i][3] == -1:
        cv2.drawContours(sep_brd,contours,i,(255,0,0),7)
```
