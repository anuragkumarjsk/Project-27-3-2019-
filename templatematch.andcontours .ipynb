```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
flock1=cv2.imread('/root/Desktop/python/flock1.jpeg')
flock1_face=cv2.imread('/root/Desktop/python/flock1_face.jpeg')

flock1=cv2.cvtColor(flock1,cv2.COLOR_BGR2GRAY)
kernel=np.ones((2,2),dtype=np.uint8)
flock1=cv2.erode(flock1,kernel,iterations=1)
plt.imshow(flock1,cmap='gray')
```

```python
plt.imshow(flock1_face)
```

```python
methods=['cv2.TM_CCOEFF','cv2.TM_CCOEFF_NORMED','cv2.TM_CCORR','cv2.TM_CCORR_NORMED','cv2.TM_SQDIFF','cv2.TM_SQDIFF_NORMED']
```

```python
for meth in methods:
    method=eval(meth)
    res=cv2.matchTemplate(flock1,flock1_face,method)
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
```

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
flock1=cv2.imread('/root/Desktop/python/flock2.jpeg',0)
```

```python
plt.imshow(flock1,cmap='gray')
```

```python
contours,hierarchy=cv2.findContours(flock1,cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
```

```python
len(hierarchy[0])
```

```python
type(contours)
```

```python
external_contours=np.zeros(flock1.shape)
```

```python
for i in range(len(contours)):
    if hierarchy[0][i][3]!=-1:
        cv2.drawContours(external_contours,contours,i,255,5)
```

```python
plt.imshow(external_contours,cmap='gray')
```
