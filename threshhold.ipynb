```python
def img_show(img):
    fig=plt.figure(figsize=(15,15))
    aa=fig.add_subplot(111)
    aa.imshow(img,cmap='gray')
```

```python
img_show(img)
```

```python
ret,thresh12=cv2.threshold(img,115,255,cv2.THRESH_BINARY)
img_show(thresh12)
```

```python
thresh2=cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,8)
img_show(thresh2)
```

```python
blended=cv2.addWeighted(src1=thresh12,alpha=0.3,src2=thresh2,beta=0.7,gamma=0)
img_show(blended)
```
