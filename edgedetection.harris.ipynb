```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
flat_chess=cv2.imread('/root/Desktop/python/chess.jpeg')
chess=cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)
plt.imshow(chess,cmap='gray')
```

```python
real_chess=cv2.imread('/root/Desktop/python/chessboard.jpeg')
real_chess=cv2.cvtColor(real_chess,cv2.COLOR_BGR2RGB)
plt.imshow(real_chess)
```

```python
real_chess_gray=cv2.imread('/root/Desktop/python/chessboard.jpeg')
real_chess_gray=cv2.cvtColor(real_chess_gray,cv2.COLOR_BGR2GRAY)
plt.imshow(real_chess_gray,cmap='gray')
```

```python
gray=np.float32(chess)
```

```python
dest=cv2.cornerHarris(src=gray,blockSize=2,ksize=3,k=0.04)
dest=cv2.dilate(dest,None)
flat_chess[dest>0.01*dest.max()]=[255,0,0]
```

```python
plt.imshow(flat_chess)
```

```python
gray=np.float32(real_chess_gray)
```

```python
dest=cv2.cornerHarris(src=gray,blockSize=2,ksize=3,k=0.04)
dest=cv2.dilate(dest,None)
real_chess[dest>0.01*dest.max()]=[255,0,0]
```

```python
plt.imshow(real_chess)
```
