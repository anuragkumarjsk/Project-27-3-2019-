```python
import cv2
import matplotlib.pyplot as plt
#%matplotlib.inline
img=cv2.imread('/root/Desktop/python/parrot.jpeg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
cv2.imwrite('parrothue.jpg',img)
plt.imshow(img)
```

```python
import cv2
import matplotlib.pyplot as plt
#%matplotlib.inline
img1=cv2.imread('/root/Desktop/python/parrot.jpeg')
img1=cv2.resize(img1,(250,170))
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2=cv2.imread('/root/Desktop/python/parrothue.jpg')
img2=cv2.resize(img2,(250,170))

merged=cv2.addWeighted(src1=img1,alpha=0.7,src2=img2,beta=0.3,gamma=0)
plt.imshow(merged)
```

```python
# diff sizes overlay
import cv2
import matplotlib.pyplot as plt

img1=cv2.imread('/root/Desktop/python/condor.jpeg')
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2=cv2.imread('/root/Desktop/python/raven.jpeg')
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
img2=cv2.resize(img2,(122,125))
large_image=img1
small_image=img2

x_offset=0
y_offset=0

x_end=x_offset+small_image.shape[1]
y_end=y_offset+small_image.shape[0]

large_image[y_offset:y_end,x_offset:x_end]=small_image

plt.imshow(large_image)

```
