```python
methods=['cv2.TM_CCOEFF','cv2.TM_CCOEFF_NORMED','cv2.TM_CCORR','cv2.TM_CCORR_NORMED','cv2.TM_SQDIFF','cv2.TM_SQDIFF_NORMED']
```

```python
method=eval('cv2.TM_CCOEFF')
res=cv2.matchTemplate(raven,raven_face,method)
```

```python
plt.imshow(res)
```

```python
#heat map

for m in methods:
    r_copy=raven.copy()
    method=eval(m)
    res=cv2.matchTemplate(r_copy,raven_face,method)
    minvalue,maxvalue,minlocation,maxlocation=cv2.minMaxLoc(res)
    
    if m in [cv2.TM_SQDIFF_NORMED,cv2.TM_SQDIFF]:
        topleft=minlocation
    else:
        topleft=maxlocation
        
    height,width,channels=raven_face.shape
    bottomright=(topleft[0]+width,topleft[1]+height)
    cv2.rectangle(r_copy,topleft,bottomright,(225,0,0),10)
    plt.subplot(121)
    plt.imshow(res)
    plt.title('result of heatmap')
    
    plt.subplot(122)
    plt.imshow(r_copy)
    plt.title('detection of template')
    
    plt.suptitle(m)
    plt.show()
    print("/n")
    print("/n")
    
```
