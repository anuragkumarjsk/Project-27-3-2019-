```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
image=cv2.imread('/root/Desktop/python/parrot.jpeg')
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image,cmap='gray')
```

```python
edges=cv2.Canny(image,threshold1=127,threshold2=127)
plt.imshow(edges)
```

```python
edges=cv2.Canny(image,threshold1=0,threshold2=255)
plt.imshow(edges)
```

```python
med_val=np.median(image)
```

```python
med_val
```

```python
lower=int(max(0,0.7*med_val))
upper=int(min(255,1.3*med_val))
```

```python
edges=cv2.Canny(image,threshold1=lower,threshold2=upper+70)
plt.imshow(edges)
```

```python
blurred_image=cv2.blur(image,ksize=(5,5))
edges=cv2.Canny(blurred_image,threshold1=lower,threshold2=upper)
plt.imshow(edges)
```
