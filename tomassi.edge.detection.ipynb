```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
flat_chess=cv2.imread('/root/Desktop/python/chess.jpeg')
flat_chess=cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)

real_chess=cv2.imread('/root/Desktop/python/chessboard.jpeg')
real_chess=cv2.cvtColor(real_chess,cv2.COLOR_BGR2RGB)

gray_flat_chess=cv2.cvtColor(flat_chess,cv2.COLOR_RGB2GRAY)
gray_real_chess=cv2.cvtColor(real_chess,cv2.COLOR_RGB2GRAY)
```

```python
corners=cv2.goodFeaturesToTrack(gray_flat_chess,5,0.01,10)
```

```python
corners=np.int0(corners)
```

```python
for i in corners:
    x,y=i.ravel()
    cv2.circle(flat_chess,(x,y),3,(255,0,0),-1)
    
plt.imshow(flat_chess)
```

```python
corners=cv2.goodFeaturesToTrack(gray_real_chess,80,0.01,10)
```

```python
corners=np.int0(corners)
```

```python
for i in corners:
    x,y=i.ravel()
    cv2.circle(real_chess,(x,y),3,(255,0,0),-1)
    
plt.imshow(real_chess)
```
